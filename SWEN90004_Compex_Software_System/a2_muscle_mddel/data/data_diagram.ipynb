```python
# this code is only used to print diagram from csv data
## this is not a part our project 
### the model replication code is in java
### bichuan Liu

### import matplotlib and pandas for data process
import matplotlib.pyplot as plt
import pandas as pd

nutrition = pd.read_csv("data.csv")
origin = pd.read_csv("data2.csv")
```

```python
days = nutrition["DAY"]
# print(days)
origin_mass = origin["Muscle Mass"]
origin_ana = origin["Anabolic"]
origin_cat = origin["Catabolic"]

nutrition_mass = nutrition["Muscle Mass"]
nutrition_ana = nutrition["Anabolic"]
nutrition_cat = nutrition["Catabolic"]
```

```python
# draw mass diagram between original and nutrition model
plt.figure(figsize=(10, 5))
plt.plot(days, origin_mass, marker=',', label='Origin model')
plt.plot(days, nutrition_mass, marker=',', label='Nutrition model')
plt.title('Muscle Development')
plt.ylabel('Muscle Mass')
plt.xlabel('DAYS')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# draw anabolic and catabolic diagram between original and nutrition model
plt.figure(figsize=(10, 5))
plt.plot(days, origin_ana, marker=',', label='Origin model Anabolic')
plt.plot(days, nutrition_ana, marker=',', label='Nutrition model Anabolic')
plt.plot(days, origin_cat, marker=',', label='Origin model Catabolic')
plt.plot(days, nutrition_cat, marker=',', label='Nutrition model Catabolic')
plt.title('Hormones')
plt.ylabel('Hormones')
plt.xlabel('DAYS')
plt.legend()
plt.grid(True)
plt.show()
```
